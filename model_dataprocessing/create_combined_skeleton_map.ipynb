```python
import pandas as pd
from tqdm import tqdm

sentence_dataframe = pd.read_csv('../extractedbtvdataset/sentences.csv')
sentence_mapped_skeletons_root = "../extractedbtvdataset/skeletons_with_sentence_id"

# sentence_dataframe = sentence_dataframe.head(2)
```

```python

all_rows_with_selected_columns = pd.DataFrame()
for index, row in tqdm(sentence_dataframe.iterrows(), total=sentence_dataframe.shape[0]):
    video_name = row['video_name']
    sentence_id = row['sentence_id']
    skeleton_csv_path = f"{sentence_mapped_skeletons_root}/{video_name}/{sentence_id}.csv"
    skeleton_dataframe = pd.read_csv(skeleton_csv_path)
    skeleton_dataframe = skeleton_dataframe.rename(columns={'Unnamed: 0': 'frame_index_in_sentence'})
    skeleton_dataframe['sentence_id'] = sentence_id

    cols = list(skeleton_dataframe.columns)
    cols.insert(0, cols.pop(cols.index('sentence_id')))
    skeleton_dataframe = skeleton_dataframe[cols]
    all_rows_with_selected_columns = pd.concat([skeleton_dataframe, all_rows_with_selected_columns], axis=0)
```

```python
all_rows_with_selected_columns = all_rows_with_selected_columns.reset_index(drop=True)
```
